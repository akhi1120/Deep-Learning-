```python

import matplotlib.pyplot as plt

def cv2_imshow(image, cmap=None):
    """Display an image in a Jupyter/VS Code notebook."""
    plt.imshow(image, cmap=cmap)
    plt.axis("off")
    plt.show()

# Load and display a grayscale image
grayscale_image = cv2.imread('C:\\Users\\akhil\\Downloads\\Deep Learning_Projects\\images.jpg', cv2.IMREAD_GRAYSCALE)
cv2_imshow(grayscale_image, cmap="gray")
```
